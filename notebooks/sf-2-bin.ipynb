```python
import torch
from safetensors.torch import load_file

# Specify your model paths
safetensor_model_path = '../model.safetensors'  # Path to your .safetensors file
bin_model_path = '../model.bin'                   # Desired path for the .bin file

# Load the .safetensors file
pt_state_dict = load_file(safetensor_model_path, device="cpu")

# Save the loaded state dict to .bin format
torch.save(pt_state_dict, bin_model_path)
```
